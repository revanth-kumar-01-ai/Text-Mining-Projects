```python
from wordcloud import WordCloud

import matplotlib.pyplot as plt

from nltk.util import ngrams

from collections import Counter

import joblib
```

```python
tokenize = joblib.load('../artifacts/tokenize')
tokenize[:5]
```

# Single-gram visual

```python
text_for_wordcloud = " ".join(tokenize)

# Generate the word cloud
wordcloud = WordCloud(width=800, height=400, background_color="white").generate(text_for_wordcloud)

# Display the word cloud
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")  # Hide axes
plt.show()
```

# Bi-gram visual

```python
bigrams = list(ngrams(tokenize, 2))
```

```python
bigrams[:5]
```

```python
bigram_phrases = [" ".join(bigram) for bigram in bigrams]
```

```python
bigram_freq = Counter(bigram_phrases)
```

```python
wordcloud = WordCloud(width=800, height=400, background_color="white").generate_from_frequencies(bigram_freq)


plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")  # Hide axes
plt.show()
```
